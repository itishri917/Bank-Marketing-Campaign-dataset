```python
import pandas as pd
import numpy as np
```

```python
# Importing data from csv to pandas data frame

df = pd.read_csv("C:/QL/DSA Data set.csv")
```

```python
# Renaming target variable
df = df.rename(columns={'y': 'TermDeposit'})
df
```

# DATA CLEANING AND FORMATTING



```python
# Finding null values

df.info()

# Null values do not exist
```

```python
# Finding duplicate values

dup_values = df[df.duplicated(keep='first')]
```

```python
# There are 12 duplicate rows

dup_values
```

```python
# Delete duplicate rows

df=df.drop_duplicates()

df
```

```python
## Counting unique values of each column

df.nunique(axis=0)
```

```python
# To perform one hot encoding later, convert object variables to category type

cat = df.select_dtypes(include=['object'])
```

```python

for col in cat:
    df[col]=df[col].astype('category')
```

```python
df
```

```python
df['Actual'] = np.where(df['TermDeposit']=='yes' , 1, 0)
df
```

```python
# correlation matrix to see relation between continuous attributes

import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr()

f, ax = plt.subplots(figsize=(10,12))

cmap = sns.diverging_palette(200, 11, as_cmap=True)

_ = sns.heatmap(corr, square=True, ax=ax, annot=True, linewidth=0.1, cmap="Blues")

plt.title("Pearson correlation", y=1.05, size=15)


# Attributes such as nr.employed, euribor3m, emp.var.rate and pdays have a strong correlation with value od Termdeposit
```

# MY PREDICTIVE MODELING

```python
## LOGISTIC REGRESSION 

from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
```

```python
## Separate inpedendent and target variables

X = df.drop(['TermDeposit','ModelPrediction','duration','predicted'], axis=1)
```

```python
X.shape
```

```python
Y = df.TermDeposit
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.20, random_state = 42)

Y_train.replace({"no":0, "yes":1}, inplace=True)
Y_test.replace({"no":0, "yes":1}, inplace=True)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced')
model.fit(X_train, Y_train)
y_pred = model.predict_proba(X_test)
```

```python
y_pred
```

```python
print("AUC score: ", roc_auc_score(Y_test, y_pred[:,1]))
```

```python
## Plotting ROC curve for my model :


preds = y_pred[:,1]
fpr, tpr, threshold = metrics.roc_curve(Y_test, preds)
roc_auc = metrics.auc(fpr, tpr)
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
